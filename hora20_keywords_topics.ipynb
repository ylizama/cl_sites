```python
from bs4 import BeautifulSoup
import os
import glob
import csv
```

```python
def parse_htmls(path, csv_file):
    csv_file = open(csv_file, "w")
    csv_writer = csv.writer(csv_file, delimiter='\t')
    
    for filename in glob.glob(os.path.join(path, '*.html')):
        soup = BeautifulSoup(open(filename), 'html.parser')
        title  = soup.title.string
        title = title.split(' | ')[1]
        print(title)
        topic = soup.find(tid = 'ALL TOPICS')
        while topic:
            topic = topic.find_next('ul')
            word = topic.get_text()
            print('w', word)
            keys = topic.find_all('li')
            for k in keys:
                print('k', k.get_text())
            
        break;
```
