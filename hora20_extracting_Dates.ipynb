```python
def extract_dates(text):
    dates = []
    patron = re.compile('((primero|uno|dos|tres|cuatro|cinco|seis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince|dieciséis|diecisiete|dieciocho|diecinueve|veinte|veintiuno|vientidos|veintitres|veinticuatro|veinticinco|veintiséis|veintisiete|veintiocho|veintinueve|treinta|treinta y uno) de (enero|febrero|marzo|abril|mayo|junio|julio|agosto|septiembre|octubre|noviembre|diciembre) de ((mil|dos mil) (cien|doscientos|trescientos|cuatrocientos|quinientos|seiscientos|setecientos|ochocientos|novecientos)( dieci| veinti| treinta y| cuarenta y| cincuenta y| sesenta y| setenta y| ochenta y| noventa y)?( )?(uno|dos|tres|cuatro|cinco|seis|séis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince)))')
    dates = patron.findall(text)    
    dates = [x[0] for x in dates]
    
    patron3 = re.compile('((enero|febrero|marzo|abril|mayo|junio|julio|agosto|septiembre|octubre|noviembre|diciembre)? (del|de|desde|desde el) (primero|uno|dos|tres|cuatro|cinco|seis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince|dieciséis|diecisiete|dieciocho|diecinueve|veinte|veintiuno|vientidos|veintitres|veinticuatro|veinticinco|veintiséis|veintisiete|veintiocho|veintinueve|treinta|treinta y uno)( al )(primero|uno|dos|tres|cuatro|cinco|seis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince|dieciséis|diecisiete|dieciocho|diecinueve|veinte|veintiuno|vientidos|veintitres|veinticuatro|veinticinco|veintiséis|veintisiete|veintiocho|veintinueve|treinta|treinta y uno)( de )?(enero|febrero|marzo|abril|mayo|junio|julio|agosto|septiembre|octubre|noviembre|diciembre)?( de )?((mil|dos mil) (cien|doscientos|trescientos|cuatrocientos|quinientos|seiscientos|setecientos|ochocientos|novecientos)( dieci| veinti| treinta y| cuarenta y| cincuenta y| sesenta y| setenta y| ochenta y| noventa y)?( )?(uno|dos|tres|cuatro|cinco|seis|séis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince))?)')
    dates3 = patron3.findall(text)    
    dates3 = [x[0] for x in dates3]
    dates = dates + dates3
    
    patron2 = re.compile('(((pasado|próximo|siguiente) ((lunes)|(martes)|(miercoles)|(miércoles)|(jueves)|(viernes)|(sabado)|(sábado)|(domingo)))|(((lunes)|(martes)|(miercoles)|(miércoles)|(jueves)|(viernes)|(sabado)|(sábado)|(domingo)) (pasado|próximo|siguiente|que viene|primero|uno|dos|tres|cuatro|cinco|seis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince|dieciséis|diecisiete|dieciocho|diecinueve|veinte|veintiuno|vientidos|veintitres|veinticuatro|veinticinco|veintiséis|veintisiete|veintiocho|veintinueve|treinta|treinta y uno)))')
    dates2 = patron2.findall(text)    
    dates2 = [x[0] for x in dates2]
    dates = dates + dates2
 
    patron4 = re.compile('((enero|febrero|marzo|abril|mayo|junio|julio|agosto|septiembre|octubre|noviembre|diciembre) (a|hasta) (enero|febrero|marzo|abril|mayo|junio|julio|agosto|septiembre|octubre|noviembre|diciembre)( )?(de|del)?( )?((mil|dos mil) (cien|doscientos|trescientos|cuatrocientos|quinientos|seiscientos|setecientos|ochocientos|novecientos)( dieci| veinti| treinta y| cuarenta y| cincuenta y| sesenta y| setenta y| ochenta y| noventa y)?( )?(uno|dos|tres|cuatro|cinco|seis|séis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince))?)')
    dates4 = patron4.findall(text)    
    dates4 = [x[0] for x in dates4]
    dates = dates + dates4
    
    #anno a anno ok
    patron5 = re.compile('((mil|dos mil) (cien|doscientos|trescientos|cuatrocientos|quinientos|seiscientos|setecientos|ochocientos|novecientos)( dieci| veinti| treinta y| cuarenta y| cincuenta y| sesenta y| setenta y| ochenta y| noventa y)?( )?(uno|dos|tres|cuatro|cinco|seis|séis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince) (a|al) ((dos mil|mil)()?(cien|doscientos|trescientos|cuatrocientos|quinientos|seiscientos|setecientos|ochocientos|novecientos)?( dieci| veinti| treinta y| cuarenta y| cincuenta y| sesenta y| setenta y| ochenta y| noventa y)?( )?(uno|dos|tres|cuatro|cinco|seis|séis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince)))')
    dates5 = patron5.findall(text)    
    dates5 = [x[0] for x in dates5]
    dates = dates + dates5
          
    #mes dia ok y mes dia a mes dia ok
    patron7 = re.compile('((enero|febrero|marzo|abril|mayo|junio|julio|agosto|septiembre|octubre|noviembre|diciembre) (primero|uno|dos|tres|cuatro|cinco|seis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince|dieciséis|diecisiete|dieciocho|diecinueve|veinte|veintiuno|vientidos|veintitres|veinticuatro|veinticinco|veintiséis|veintisiete|veintiocho|veintinueve|treinta|treinta y uno)( (a)( )(enero|febrero|marzo|abril|mayo|junio|julio|agosto|septiembre|octubre|noviembre|diciembre) (primero|uno|dos|tres|cuatro|cinco|seis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince|dieciséis|diecisiete|dieciocho|diecinueve|veinte|veintiuno|vientidos|veintitres|veinticuatro|veinticinco|veintiséis|veintisiete|veintiocho|veintinueve|treinta|treinta y uno))?)')
    dates7 = patron7.findall(text)
    dates7 = [x[0] for x in dates7]
    dates = dates + dates7
    
    #dia de mes al dia de mes (ok)
    patron8 = re.compile('((primero|uno|dos|tres|cuatro|cinco|seis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince|dieciséis|diecisiete|dieciocho|diecinueve|veinte|veintiuno|vientidos|veintitres|veinticuatro|veinticinco|veintiséis|veintisiete|veintiocho|veintinueve|treinta|treinta y uno) (de) (enero|febrero|marzo|abril|mayo|junio|julio|agosto|septiembre|octubre|noviembre|diciembre) (al|hasta|hasta el) (primero|uno|dos|tres|cuatro|cinco|seis|siete|ocho|nueve|diez|once|doce|trece|catorce|quince|dieciséis|diecisiete|dieciocho|diecinueve|veinte|veintiuno|vientidos|veintitres|veinticuatro|veinticinco|veintiséis|veintisiete|veintiocho|veintinueve|treinta|treinta y uno) (de) (enero|febrero|marzo|abril|mayo|junio|julio|agosto|septiembre|octubre|noviembre|diciembre))')
    dates8 = patron8.findall(text)
    dates8 = [x[0] for x in dates8]
    dates = dates + dates8
    return dates

#extract_dates("habiamos buscado de del cinco de julio al diez de agosto comied bk  ")
```

```python
import datetime
for filename in glob.glob(os.path.join(DIR_DATA, '*.rtf')):
    with open(filename) as f:
        key = filename[:-4].replace(DIR_DATA, '')        
        if key in data.keys():            
            text = f.read()
            dates = extract_dates(text.lower())
            if('hoy' in text and data[key]['date'] is not None):                
                dates.append('hoy')
            if('ayer' in text and data[key]['date'] is not None):
                dates.append('ayer')
            if('mañana' in text and data[key]['date'] is not None):
                dates.append('mañana')
            if len(dates) > 0:
                print (dates)
            data[key]['dates'] = dates
```
